```python

import pandas as pd
import numpy as np

# 1. LOAD DATASET
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("--- TASK 1: DATASET ANALYSIS ---\n")

# 2. DISPLAY FIRST AND LAST RECORDS
print("Showing first 5 records:")
display(df.head())
print("\nShowing last 5 records:")
display(df.tail())

# 3. DATA TYPES AND NULL VALUES
print("\n--- Data Structure & Info ---")
df.info()

# 4. STATISTICAL SUMMARY
print("\n--- Statistical Summary (Numerical) ---")
display(df.describe())

# 5. UNIQUE VALUES IN CATEGORICAL COLUMNS
print("\n--- Categorical Data Distribution ---")
categorical_cols = ['Survived', 'Pclass', 'Sex', 'Embarked']
for col in categorical_cols:
    print(f"\nUnique values in {col}:")
    print(df[col].value_counts())

# 6. DATA QUALITY OBSERVATIONS
print("\n--- Data Quality Report ---")
missing_data = df.isnull().sum()
print("Missing values per column:")
print(missing_data[missing_data > 0])

# Target Variable Analysis
survive_rate = df['Survived'].mean() * 100
print(f"\nSurvival Rate (Target Balance): {survive_rate:.2f}% survived / {100-survive_rate:.2f}% deceased")

# 7. ML SUITABILITY SUMMARY
print("\n[CONCLUSION]")
print("- Target Variable: 'Survived' (Binary Classification)")
print("- Features: Mixture of Numerical (Age, Fare) and Categorical (Sex, Pclass)")
print("- Suitability: High, but requires handling missing values in 'Age' and 'Cabin'.")
```
